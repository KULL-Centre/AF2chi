```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import glob
```

```python
def load_rmse_from_json(file_loc_list,order=['RMSE_C_bootstrap','RMSE_N_bootstrap','RMSE_HAHB_bootstrap']):
    rmse_collected={}
    for idx,elem in enumerate(file_loc_list):
        with open(elem,'r') as f:
            rmse_collected[order[idx]]= json.load(f)

    return rmse_collected
```

```python
def plot_violin(title, xlabel,dictionary_label_data,hps_value,pltboundaries=[0,3]):
    """
    Plots violin plots for the given lists.
    
    Parameters:
    title (str): The title of the plot.
    ylabel (str): The label for the y-axis.
    *args: An arbitrary number of lists, each containing numerical values.
    """

    # Convert data to long-form DataFrame for Seaborn
    df = pd.DataFrame({key: pd.Series(value) for key, value in dictionary_label_data.items()})
    df = df.melt(var_name='List', value_name='Value')
    # Create the violin plot
    fig,ax= plt.subplots(1,1,figsize=(4.5, 6))
    sns.boxplot(x='List', y='Value', data=df, ax=ax, color='k',zorder=10)

    sns.stripplot(x='List', y='Value', data=df, jitter=True, ax=ax, zorder=-10,color='gray',edgecolor='white',linewidth=0.8, size=6)

    
    ax.axhline(y=hps_value, color='r', linestyle='--', linewidth=2,label='full HSP')
    ax.set_title(title,fontsize=16)
    ax.legend()

    ax.set_ylim(pltboundaries[0],pltboundaries[1])
    ax.set_ylabel('RMSE',fontsize=16)
    ax.set_xlabel(xlabel,fontsize=16)
```

```python
def plot_dict_comparison(dict1, dict2,hsp_value,af2chi_value,title='BTPI HA_CA_CB_HB',y_limits=[0,2]):
    """
    Plots a grouped boxplot comparing the distributions
    of common keys between two dictionaries.
    
    Parameters:
    dict1 (dict): First dictionary with column names as keys and lists of values.
    dict2 (dict): Second dictionary with column names as keys and lists of values.
    """
    # Find common columns and maintain the order from dict1
    common_keys = [key for key in dict1.keys() if key in dict2]
    if not common_keys:
        print("No common columns to compare.")
        return
    
    # Prepare data for boxplot, removing NaN values
    data1 = [np.array(dict1[key])[~np.isnan(dict1[key])] for key in common_keys]
    data2 = [np.array(dict2[key])[~np.isnan(dict2[key])] for key in common_keys]
    
    # Box plot setup
    fig, ax = plt.subplots(figsize=(6, 6))
    
    positions1 = np.arange(len(common_keys)) * 2  # Positions for dict1
    positions2 = positions1 + 1  # Positions for dict2
    
    ax.boxplot(data1, positions=positions1, widths=0.6, patch_artist=True, boxprops=dict(facecolor='#56b4e9'), medianprops=dict(color='black'))
    ax.boxplot(data2, positions=positions2, widths=0.6, patch_artist=True, boxprops=dict(facecolor='#009e74'), medianprops=dict(color='black'))
    
    ax.axhline(y=hsp_value, color='#56b4e9', linestyle='--', linewidth=2)
    ax.axhline(y=af2chi_value, color='#009e74', linestyle='--', linewidth=2)

    # Labels and formatting
    ax.set_xlabel('n_pdbs /n_bootstraps',fontsize=17)
    ax.set_ylabel('RMSE [Hz]',fontsize=17)
    ax.set_title(title,fontsize=17)
    ax.set_xticks((positions1 + positions2) / 2)
    ax.set_xticklabels(common_keys)
    ax.set_ylim(y_limits[0],y_limits[1])
    ax.legend(["HSP ensemble", "AF2chis"], loc='upper right',fontsize=17)
```
